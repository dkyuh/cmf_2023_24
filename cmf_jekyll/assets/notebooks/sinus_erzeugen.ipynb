```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

## Sinuswelle erzeugen

Allgemeine Formel im kontinuierlichen Bereich: $$y(t) = A \cdot \sin(2 \pi t f + \phi)$$ wobei:
- $A$ die Amplitude ist
- $f$ die Frequenz (in Hz)
- $t$ die Zeit
- $\phi$ die Phase (Verschiebung auf der x-Achse)

```python
sr = 50 # sampling rate
length = 2 # sec
f = 1 # frequenz in Hz
phase = 0 # phase in radianten
amp = 1 # amplitude

# erzeugen der zeitachse mit linspace
t = np.linspace(0, length, num=int(sr * length))
# erzeugen der sinuswelle
y = np.sin(2 * np.pi * t * f - phase) * amp

# plot
plt.plot(t, y, '-o')
plt.xlabel('t (Zeit)')
plt.ylabel('Amplitude')
plt.show()
```

Wenn wir eher zeitdiskret "denken" möchten, können wir die Zeit durch eine Zahlenfolge $n$ ersetzen, die von 0-N geht, wobei $N$ die Anzahl der Samples im Signal ist:

```python
sr = 50
length = 2 # sec
N = int(sr * length) # anzahl der samples im signal

f = 3
# phase = np.pi / 2
phase = 0
amp = 0.5

n = np.arange(0, N)
y = np.sin(2 * np.pi * n / sr * f - phase) * amp

plt.plot(t, y, '-o')
plt.xlabel('t (Zeit)')
plt.ylabel('Amplitude')
plt.ylim(-1, 1)
plt.show()
```

```python
# hoerbarer Bereich

sr = 44100
length = 2 # sec
N = int(sr * length)

f = 300
phase = 0
amp = 0.5

n = np.arange(0, N)
y = np.sin(2 * np.pi * n / sr * f - phase) * amp

plt.plot(n / sr, y)
plt.xlabel('t (Zeit)')
plt.ylabel('Amplitude')
plt.ylim(-1, 1)
plt.show()

display(Audio(y, rate=sr))
```
